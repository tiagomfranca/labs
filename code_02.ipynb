# 1a parte

No espaço de código abaixo, plote as 3 imagens, uma ao lado da outra. Ainda, plote o histograma de cada canal de cor (R - Vermelho, G - Verde e B - Azul) abaixo da respectiva imagem.

```
img_plenio = Image.open('/content/drive/My Drive/johannes-plenio-RwHv7LgeC7s-unsplash.jpg')
img_testa = Image.open('/content/drive/My Drive/claudio-testa--SO3JtE3gZo-unsplash.jpg')
img_zindel = Image.open('/content/drive/My Drive/bailey-zindel-NRQV-hBF10M-unsplash.jpg')

img_plenio_arr = array(img_plenio)
img_testa_arr = array(img_testa)
img_zindel_arr = array(img_zindel)

plt.figure(figsize=(20,4))

plt.subplot(341)
plt.imshow(img_plenio)
plt.axis('off')
plt.title('Imagem Plenio')

r_plenio = img_plenio_arr[:,:,0]
g_plenio = img_plenio_arr[:,:,1]
b_plenio = img_plenio_arr[:,:,2]

plt.subplot(342)
plt.hist(r_plenio.ravel(), color='red')

plt.subplot(343)
plt.hist(g_plenio.ravel(), color='green')

plt.subplot(344)
plt.hist(b_plenio.ravel(), color='blue')

plt.subplot(345)
plt.imshow(img_testa)
plt.axis('off')
plt.title('Imagem Testa')

r_testa = img_testa_arr[:,:,0]
g_testa = img_testa_arr[:,:,1]
b_testa = img_testa_arr[:,:,2]

plt.subplot(346)
plt.hist(r_testa.ravel(), color='red')

plt.subplot(347)
plt.hist(g_testa.ravel(), color='green')

plt.subplot(348)
plt.hist(b_testa.ravel(), color='blue')

plt.subplot(349)
plt.imshow(img_zindel)
plt.axis('off')
plt.title('Imagem Zindel')

r_zindel = img_zindel_arr[:,:,0]
g_zindel = img_zindel_arr[:,:,1]
b_zindel = img_zindel_arr[:,:,2]

plt.subplot(3, 4, 10)
plt.hist(r_zindel.ravel(), color='red')

plt.subplot(3,4, 11)
plt.hist(g_zindel.ravel(), color='green')

plt.subplot(3, 4, 12)
plt.hist(b_zindel.ravel(), color='blue')

plt.show()
```

Vamos gerar algumas imagens preto e branco, utilizando limiares diferentes:

*  45
*  100
*  145
*  200
*  245

Escreva o código no espaço abaixo.
